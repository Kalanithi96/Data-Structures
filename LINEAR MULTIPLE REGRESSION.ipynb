```python
df.drop_duplicates(inplace= True)
df.info()
```

```python
df.drop(["car_ID","CarName","fueltype","aspiration","doornumber","carbody","carlength","carwidth","enginesize","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem"],axis=1,inplace=True)
df.info()
```

```python
from sklearn.model_selection import train_test_split

x_df=df.drop("price",axis=1)
y_df=df[["price"]]

Xtrain, Xtest, Ytrain, Ytest = train_test_split(x_df, y_df, test_size=0.22, random_state=42, shuffle=True)
```

```python
Xtrain.info()
Ytrain.info()
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt

reg = linear_model.LinearRegression()

reg.fit(Xtrain, Ytrain)

print('Coefficients: ', reg.coef_)
 
print('Variance score: {}'.format(reg.score(Xtest, Ytest)))



plt.scatter(reg.predict(Xtrain), reg.predict(Xtrain) - Ytrain, color = "green", s = 10, label = 'Train data')
 
plt.scatter(reg.predict(Xtest), reg.predict(Xtest) - Ytest, color = "blue", s = 10, label = 'Test data')
```

```python
import numpy as np
```

```python
X = np.array(Xtrain)
print(X)
```

```python
Y = np.array(Ytrain)
print(Y)
```

```python
X2=X
X3=np.array([1]*X2.shape[0],ndmin=2)
X3=X3.reshape(X2.shape[0],1)
X2=np.concatenate((X3,X2),axis=1)
print(X2)
```

```python
coef = np.matmul( np.linalg.inv(np.matmul(np.transpose(X2),X2)) , np.matmul(np.transpose(X2),Y) )

print(coef)
```

```python
X4=Xtest
X5=np.array([1]*X4.shape[0],ndmin=2)
X5=X5.reshape(X4.shape[0],1)
X4=np.concatenate((X5,X4),axis=1)
print(X4)
```

```python
Y_pred= np.matmul(X4,coef)
print(Y_pred)
```

```python
mean_y = np.mean(Ytest)
r_score =1 - ( np.sum((Y_pred-Ytest)**2) / np.sum((Ytest-mean_y)**2) )
print(r_score)
```
